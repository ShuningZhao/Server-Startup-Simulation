```python
import pandas as pd
from scipy.stats import kstest
from scipy.stats import expon
from scipy.stats import gamma
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
arrival = pd.read_csv('arrival.csv')
interarrival = arrival['Interarrival']
```

```python
# Kolmogorov-Smirnov test for interarrival times
kstest(interarrival*0.35, 'expon')
```

```python
x = np.linspace(0,35, 5000)

fig, ax = plt.subplots(1, 1)
sns.distplot(interarrival, label='Interarrival times')
ax.plot(x, expon.pdf(x, scale = 2.8),'r-', lw=3, alpha=0.6, label='Exponential pdf')
plt.legend();
```

```python
service = pd.read_csv('service.csv')
service_time = arrival['Service']
```

```python
# Kolmogorov-Smirnov test for service times
kstest(service_time, 'gamma', args=(3, 0, 1))
```

```python
fig, ax = plt.subplots(1, 1)
sns.distplot(service_time, label='Service times')
ax.plot(x, gamma.pdf(x, a = 3, scale = 1),'r-', lw=3, alpha=0.6, label='Gamma pdf')
plt.legend();
```
